```python
age_mean = X_train['Age'].mean()
X_train['age_mean_imputer'] = X_train['Age'].fillna(age_mean)
X_train
X_test['age_mean_imputer'] = X_test['Age'].fillna(age_mean)
X_test
```

```python
embarked_imputor = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
embarked_imputor.fit(X_train[['Embarked']]) 
X_train['Embarked'] = embarked_imputor.transform(X_train[['Embarked']]).ravel()
X_train
```

# Categorical Value imputation

```python
cabin_imputor = SimpleImputer(missing_values=np.nan,strategy='constant', fill_value='missing', add_indicator=True)
cabin_imputor.fit(X_train[['Cabin']])
X_train[['Cabin','Cabin_missing_indicator']] = cabin_imputor.transform(X_train[['Cabin']])
```

```python
X_train
```

# Outlier Detection

```python
mean_of_age = X_train['Age'].mean()
sd_of_age = X_train['Age'].std()

X_train['Zscore_Age'] = (X_train['Age'] - mean_of_age)/sd_of_age

X_train
```

```python
mean_of_fare = X_train['Fare'].min()
sd_of_fare = X_train['Fare'].std()
X_train['Zscore_fare'] = (X_train['Fare'] - mean_of_fare)/sd_of_fare
X_train
```

# Outliers with IQR

```python
outliers = X_train[abs(X_train['Zscore_Age']) >3]
print(len(outliers))
display(outliers)
```

```python


age_Q1  = X_train['Age'].quantile(0.25)
age_Q3 = X_train['Age'].quantile(0.75)

age_IQR = age_Q3 - age_Q1

minimum_age = age_Q1 - 1.5 * age_IQR
maximum_age = age_Q3 + 1.5 * age_IQR

print(minimum_age, maximum_age)
```

```python
age_outlier_IQR = X_train[(X_train['Age']>maximum_age) | (X_train['Age']<minimum_age)]

length_of_outliers = len(age_outlier_IQR)
print(length_of_outliers)
display(age_outlier_IQR)
```

```python
X_train = X_train[abs(X_train['Zscore_Age']<=3)]
X_train
```

```python
fare_Q1  = X_train['Fare'].quantile(0.25)
fare_Q3 = X_train['Fare'].quantile(0.75)

fare_IQR = fare_Q3 - fare_Q1

minimum_fare = fare_Q1 - 1.5 * fare_IQR
maximum_fare = fare_Q3 + 1.5 * fare_IQR

print(minimum_fare, maximum_fare)
```

```python
fare_outlier_IQR = X_train[(X_train['Fare']>maximum_fare) | (X_train['Fare']<minimum_fare)]

length_of_outliers = len(fare_outlier_IQR)
print(length_of_outliers)
display(fare_outlier_IQR)
```

```python
minimum_zscore_fare = X_train[abs(X_train['Zscore_fare']<=3)]['Fare'].min()
print(minimum_zscore_fare)

maximum_zscore_fare = X_train[abs(X_train['Zscore_fare']>=3)]['Fare'].max()
print(maximum_zscore_fare)
```

```python
X_train['Fare']= X_train['Fare'].clip(minimum_fare , maximum_fare)
maximum = X_train['Fare'].max()
print(maximum)
```
